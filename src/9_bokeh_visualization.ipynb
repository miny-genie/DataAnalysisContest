```python
dates = pd.DataFrame(pd.date_range("2022-01-01", "2022-12-31", freq='D'))

group_tailgate = preprocessed_df_tailgate.groupby('발생일')
group_illegal_u_turn = preprocessed_df_illegal_u_turn.groupby('발생일')
group_illegal_left_turn = preprocessed_df_illegal_left_turn.groupby('발생일')
group_signal_violation = preprocessed_df_signal_violation.groupby('발생일')
group_wrong_way_driving = preprocessed_df_wrong_way_driving.groupby('발생일')
group_stop_line_violation = preprocessed_df_stop_line_violation.groupby('발생일')
group_center_line_violation = preprocessed_df_center_line_violation.groupby('발생일')
group_designated_lane_violation = preprocessed_df_designated_lane_violation.groupby('발생일')
group_improper_lane_change = preprocessed_df_improper_lane_change.groupby('발생일')

all_groupby = [
    ('꼬리물기', group_tailgate),
    ('불법 유턴', group_illegal_u_turn),
    ('불법 좌회전', group_illegal_left_turn),
    ('신호 위반', group_signal_violation),
    ('역주행', group_wrong_way_driving),
    ('정지선 침범', group_stop_line_violation),
    ('중앙선 침범', group_center_line_violation),
    ('지정 차로 위반', group_designated_lane_violation),
    ('진로 변경 방법 위반', group_improper_lane_change)
]


for column_name, GROUP_BY in all_groupby:
    dlist, llist = [], []
    
    for d, l in GROUP_BY:
        dlist.append(d)
        llist.append(len(l))
    
    tmp_df = pd.DataFrame({
        '발생일': dlist,
        column_name: llist
    })
    
    tmp_df.index = pd.to_datetime(tmp_df['발생일'])
    tmp_df = tmp_df.drop('발생일', axis=1)
    
    dates = pd.merge(dates, tmp_df, how='left', left_on = 0, right_on = '발생일')

# 데이터 프레임 NaN값 0으로 변경하기
dates = dates.fillna(0)

# 데이터 프레임 컬럼명 변경하기
dates = dates.rename(columns = {0: '발생일'})

# 데이터 프레임 출력
dates
```

```python
# CSV에서 직접 확인하기 위해 저장
path_folder = 'C:/Users/kjmin/Desktop/desktop/대학/4-2/Jupyter_Notebook/preprocessed_csv/'

dates.to_csv(
    path_or_buf = path_folder + "GRAPH_GROUP_BY_DATAFRAME.csv",
    encoding="utf-8-sig"
)
```

```python
p = figure(
    title='2022년 민원 신고 데이터(단위: 건)',
    height = 700,
    width  = 1900,
    x_axis_type='datetime',
    tools='pan, wheel_zoom, box_zoom, reset, crosshair'
)

names  = ['꼬리물기', '불법 유턴', '불법 좌회전', '신호 위반', '역주행', '정지선 침범', '중앙선 침범', '지정 차로 위반', '진로 변경 방법 위반']
colors = ['#FF0000', '#00FF00', '#0000FF', '#AA0000', '#00AA00', '#0000AA', '#550000', '#005500', '#000055']

for i in range(9):
    p.line(
        dates['발생일'],            # 그래프의 x축으로 지정할, 데이터프레임의 column 선택
        dates[names[i]],            # 선 그래프로 그릴 column 이름, names 리스트 기준으로 지정
        legend_label = names[i],    # 범례 이름 names 리스트에서 지정
        line_width = 1,             # 선 두께 지정
        line_color = colors[i],     # 선 색상 colors 리스트에서 지정
        line_alpha = 1              # 선 투명도 지정
    )

    p.circle(
        dates['발생일'],            # 그래프의 x축으로 지정할, 데이터프레임의 column 선택
        dates[names[i]],            # 선 그래프로 그릴 column 이름, names 리스트 기준으로 지정
        legend_label = names[i],    # 범례 이름 names 리스트에서 지정
        size = 8,                   # 원 크기 지정
        fill_color = colors[i],     # 원 색상 colors 리스트에서 지정
        fill_alpha = 0.25,          # 원 투명도 지정
        line_color = colors[i],     # 원 테두리 색상 colors 리스트에서 지정
        line_alpha = 0.5            # 원 테두리 투명도 지정
    )

# 범례 설정
p.legend.location = 'top_left'
p.legend.click_policy = 'hide'

# 폰트 설정
font_large, font_medium, font_small = '16pt', '12pt', '10pt'
p.xaxis.major_label_text_font_size = font_small
p.yaxis.major_label_text_font_size = font_small
p.legend.label_text_font_size = font_small

p.title.text_font_size = font_large

# 그래프 그리기
show(p)
```
